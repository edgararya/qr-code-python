```python
import qrcode
from IPython.display import display # Ini library khusus buat nampilin gambar di Colab
```

```python
# Data QR
data = "https://github.com/edgararya"
```

```python
# Buat QR
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)
qr.add_data(data)
qr.make(fit=True)
```

```python
# Bikin Gambar
img = qr.make_image(fill_color="black", back_color="white")
```

```python
# TAMPILKAN LANGSUNG (Fitur asik Colab)
display(img)
```

```python
# Simpan file (ini akan tersimpan di folder sementara Colab)
img.save("hasil_qr.png")
```
